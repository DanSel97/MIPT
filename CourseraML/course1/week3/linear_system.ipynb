```python
%matplotlib inline
from matplotlib import pylab as plt
import numpy as np
import scipy as sc
from scipy import linalg
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)
xnew = np.arange(1, 15, 0.1)
ynew = f(xnew)
a = np.array([float(1), float(15)])
b = f(a)
A1 = [[float(1), float(1)], [float(1), float(15)]]
A = np.array(A1)
ans = sc.linalg.solve(A, b)
def y(x):
    return ans[0] + x*ans[1]
ynew1 = y(xnew)
plt.figure (1)
plt.plot(xnew, ynew)
plt.figure (1)
plt.plot(xnew, ynew1)
plt.show()
a = np.array([float(1), float(8), float(15)])
b = f(a)
A1 = [[float(1), float(1), float(1)], [float(1), float(8), float(8 ** 2)], [float(1), float(15), float(15 ** 2)]]
A = np.array(A1)
ans = sc.linalg.solve(A, b)
def y1(x):
    return ans[0] + x*ans[1] + (x ** 2) * ans[2]
ynew1 = y1(xnew)
plt.figure (1)
plt.plot(xnew, ynew)
plt.figure (1)
plt.plot(xnew, ynew1)
plt.show()
a = np.array([float(1), float(4), float(10), float(15)])
b = f(a)
A1 = [[float(1), float(1), float(1), float(1)], [float(1), float(4), float(4 ** 2), float(4 ** 3)], [float(1), float(10), float(10 ** 2), float(10 ** 3)], [float(1), float(15), float(15 ** 2), float(15 ** 3)]]
A = np.array(A1)
ans = sc.linalg.solve(A, b)
def y2(x):
    return ans[0] + x*ans[1] + (x ** 2) * ans[2] + (x ** 3) * ans[3]
ynew1 = y2(xnew)
plt.figure (1)
plt.plot(xnew, ynew)
plt.figure (1)
plt.plot(xnew, ynew1)
plt.show()
file_obj = open('submission-2.txt', 'w')
string = str(ans[0]) + ' ' + str(ans[1]) + ' ' + str(ans[2]) + ' ' + str(ans[3])
file_obj.write(string)
file_obj.close()
```
