```python
import numpy as np
from scipy.stats import expon
import matplotlib.pyplot as plt
%matplotlib inline
```

# Экспоненциальное распределение 


- ###### Обозначается, как: $$ {\displaystyle \mathrm {Exp} (\lambda )} $$

- ###### Определение (для θ > 0 и x ≥ 0): $$ f(x)=\dfrac{1}{\theta} e^{-x/\theta} $$

- ###### Дисперсия $$ {\displaystyle \lambda ^{-2}} $$

- ###### Математическое ожидание: $$ {\displaystyle \lambda ^{-1}} $$

- ###### Медиана: $$ {\displaystyle \ln(2)/\lambda } $$

- ###### Мода: $$ 0 $$

[Exponential distribution](https://en.wikipedia.org/wiki/Exponential_distribution)

Экспоненциальное (или показательное) распределение — абсолютно непрерывное распределение, моделирующее время между двумя последовательными свершениями одного и того же события.

Cначала сгенерируем выборку объёма 1000, построим гистограмму выборки.


```python
values = expon.rvs(size=1000)

plt.hist(values, normed=True)

plt.title("Exponential distribution")
plt.xlabel("X (values)")
plt.ylabel("Frequency")
plt.show()
```

Нарисуем поверх неё теоретическую плотность распределения различных случайных величин:

```python
values = expon.rvs(size=1000)
x = np.linspace(-5, 5, 1000)

plt.hist(values, normed=True) # create histogram

plt.plot(x, expon.pdf(x), color='r') # Display the probability density function

plt.title("Exponential distribution")
plt.xlabel("X (values)")
plt.ylabel("Frequency")
plt.show() # And compare the histogram
```

```python
values = expon.rvs(size=1000)
x1 = np.linspace(-10, 10, 1000)

plt.hist(values, normed=True) # create histogram

plt.plot(x1, expon.pdf(x1), color='r') # Display the probability density function

plt.title("Exponential distribution")
plt.xlabel("X (values)")
plt.ylabel("Frequency")
plt.show() # And compare the histogram
```

```python
values = expon.rvs(size=1000)
x2 = np.linspace(-40, 40, 1000)

plt.hist(values, normed=True) # create histogram

plt.plot(x2, expon.pdf(x2), color='r') # Display the probability density function

plt.title("Exponential distribution")
plt.xlabel("X (values)")
plt.ylabel("Frequency")
plt.show() # And compare the histogram
```

```python
# n = 5
loc, scale = expon.fit(x, floc=0) # calculate required parameters

print('Median is %f' % expon.median(loc=loc, scale=scale))
print('Mean is %f' % expon.mean(loc=loc, scale=scale))
print('Standard deviation is %f' % expon.std(loc=loc, scale=scale))
```

```python
# n = 10
loc, scale = expon.fit(x1, floc=0) # calculate required parameters

print 'Median is %f' % expon.median(loc=loc, scale=scale)
print 'Mean is %f' % expon.mean(loc=loc, scale=scale)
print 'Standard deviation is %f' % expon.std(loc=loc, scale=scale)
```

```python
# n = 40
loc, scale = expon.fit(x2, floc=0) # calculate required parameters

print 'Median is %f' % expon.median(loc=loc, scale=scale)
print 'Mean is %f' % expon.mean(loc=loc, scale=scale)
print 'Standard deviation is %f' % expon.std(loc=loc, scale=scale)
```

### Вывод

Исходя из полученных данных можно сделать вывод о том, что точность апроксимации распределения выборочных средних нормальным с ростом n увеличивается. А разница между полученными распределениями при различных значениях n заключается в росте средних значений за счет роста выборки и сглаживания краев гистограммы. Что говорит о том, что наиболее эффективными являются средние значения, поскольку они достаточно сглаженны или приближенны, при этом дальнейшее увеличение выборки не дает ощутимого эффекта и только сужает отрезок.
