```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm, norm
%matplotlib inline

##
s = 1.0
x = np.linspace(lognorm.ppf(0.01, s),lognorm.ppf(0.99, s), 100)

# Display cumulative distribution function  
cdf = lognorm.cdf(x,s) 
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
##
s = 1.0

#Display the probability density function
x = np.linspace(lognorm.ppf(0.01, s),lognorm.ppf(0.99, s), 100)
plt.plot(x, lognorm.pdf(x, s),'r-', lw=5, alpha=0.6, label='lognorm pdf')

#Generate random numbers
r = lognorm.rvs(s, size=1000)

#Compare the histogram
plt.hist(r, normed=True, histtype='stepfilled', alpha=0.2)
plt.xlabel('$x$')
plt.show()
##
import pylab as P
Nsize = 2
s = 1.0
mu = 1
# generate 1000 samples volume 2
arr = np.zeros(1000)
for i in range(1000):
    r = lognorm.rvs(s, size = Nsize)
    #sample means
    arr[i] = np.average(r)

#mathematical expectation and variance
Ex = np.exp(s*s/2)
Dx = ((np.exp(s*s)-1)*np.exp(s*s)/Nsize)
#display the appropriate density of the normal distribution
x = np.linspace(-3,5,100)
norm_rv = norm(loc=Ex, scale=np.sqrt(Dx))
norm_pdf = norm_rv.pdf(x)
P.plot(x, norm_pdf, color = 'red', linewidth=2.0)

P.hist(arr,50, normed=True)   
P.xlabel('$x$')
P.ylabel('$f(x)$')
P.show()   
##
import pylab as P
Nsize = 2
s = 1.0
mu = 1
# generate 1000 samples volume 2
arr = np.zeros(1000)
for i in range(1000):
    r = lognorm.rvs(s, size = Nsize)
    #sample means
    arr[i] = np.average(r)

#mathematical expectation and variance
Ex = np.exp(s*s/2)
Dx = ((np.exp(s*s)-1)*np.exp(s*s)/Nsize)
#display the appropriate density of the normal distribution
x = np.linspace(-3,5,100)
norm_rv = norm(loc=Ex, scale=np.sqrt(Dx))
norm_pdf = norm_rv.pdf(x)
P.plot(x, norm_pdf, color = 'red', linewidth=2.0)

P.hist(arr,50, normed=True)   
P.xlabel('$x$')
P.ylabel('$f(x)$')
P.show()   

##
import pylab as P
Nsize = 5
s = 1.0
mu = 1
# generate 1000 samples volume 5
arr = np.zeros(1000)
for i in range(1000):
    r = lognorm.rvs(s, size = Nsize)
    #sample means
    arr[i] = np.average(r)

#mathematical expectation and variance
Ex = np.exp(s*s/2)
Dx = ((np.exp(s*s)-1)*np.exp(s*s)/Nsize)

#display the appropriate density of the normal distribution
x = np.linspace(-3,5,100)
norm_rv = norm(loc=Ex, scale=np.sqrt(Dx))
norm_pdf = norm_rv.pdf(x)
P.plot(x, norm_pdf, color = 'red', linewidth=2.0)
P.hist(arr,50, normed=True) 
P.xlabel('$x$')
P.ylabel('$f(x)$')
P.show()     
##
import pylab as P
Nsize = 15
s = 1.0
mu = 1
# generate 1000 samples volume 15
arr = np.zeros(1000)
for i in range(1000):
    r = lognorm.rvs(s, size = Nsize)
    #sample means
    arr[i] = np.average(r)

#mathematical expectation and variance
Ex = np.exp(s*s/2)
Dx = ((np.exp(s*s)-1)*np.exp(s*s)/Nsize)

#display the appropriate density of the normal distribution
x = np.linspace(-3,5,100)
norm_rv = norm(loc=Ex, scale=np.sqrt(Dx))
norm_pdf = norm_rv.pdf(x)
P.plot(x, norm_pdf, color = 'red', linewidth=2.0)
P.hist(arr,50, normed=True) 
P.xlabel('$x$')
P.ylabel('$f(x)$')
P.show()  
```
