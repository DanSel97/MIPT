```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm, norm
from scipy.stats import pareto
import pylab as P
%matplotlib inline
```

```python
b = 3
fig, ax = plt.subplots(1, 1)
x = np.linspace(pareto.ppf(0.01, b), pareto.ppf(0.99, b), 100) 
ax.plot(x, pareto.pdf(x, b), 'r-', lw=5, alpha=0.6, label='pareto pdf')    
 
cdf = pareto.cdf(x, b) #cumulative distribution function - теоретическая плотность распределения#
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')

r = pareto.rvs(b, size=1000) #генерируем 1000 случайных чисел из распределения Парето#

ax.hist(r, b, normed=True, histtype='stepfilled', alpha=0.6) #сравниваем с гистограммой#
plt.show()
```

```python
x = np.linspace(0, 0, num=1000, endpoint=True)
N = 5 #Объём выборки - 5#
for i in range(1000):
    r = pareto.rvs(b, size=5)
    x[i] = np.average(r)
EX = b/(b - 1) #теоретическое матожидание распределения Парето, https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%9F%D0%B0%D1%80%D0%B5%D1%82%D0%BE#
DX = b/((b - 2)*((b - 1)**2))/N #теоретическая дисперсия распределения Парето#
y = np.linspace(1,5,100)
norm_rv = norm(loc=EX, scale=np.sqrt(DX)) #Генерируем нормальное распределение с вычисленными параметрами#
norm_pdf = norm_rv.pdf(y)
P.plot(y, norm_pdf, color = 'red', linewidth=2.0)

P.hist(x,50, normed=True)   #Генерируем гистограмму для сравнения, гистограмма лучше похожа на нормальное распределение#
P.xlabel('$x$')
P.ylabel('$f(x)$')
P.show()   


```
