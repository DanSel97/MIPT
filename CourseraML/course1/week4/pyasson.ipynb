$$ \sigma = \lambda, \mu = \lambda $$

```python
import matplotlib as plt
import numpy as np
import math as mt
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from scipy.stats import poisson
%matplotlib inline
```

```python
fig, ax = plt.subplots(1, 1)
lam = 0.6
x = np.arange(poisson.ppf(0.01, lam), poisson.ppf(0.99, lam))
cdf = poisson.cdf(x,lam) 
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
lam = 5
s = np.random.poisson(lam, 1000) #генерируем 1000 чисел из распределения Пуассона#
n, bins, patches = plt.hist(s, 14, normed=True)
sigma = lam
mu = lam
y = mlab.normpdf(bins, mu, sigma)
plt.plot(bins, y, 'r--')
plt.xlabel('Smarts')
plt.ylabel('Probability')
lam = 10
sigma = lam
mu = lam
plt.show()


```

```python
a = np.random.poisson(10, size=(1000, 5)) #здесь генерируем 1000 строк и 5 стобцов из распределения Пуассона#
b = np.average(a, axis=1) #считаем средние по строкам#
n, bins, patches = plt.hist(b, 14, normed=True)
y = mlab.normpdf(bins, mu, sigma)
plt.plot(bins, y, 'r--')
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.show()
```
